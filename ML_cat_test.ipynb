```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

X = catDisease.iloc[:,3:]
print(X)
y = catDisease['질병명']
print(y)

tree_clf = DecisionTreeClassifier(max_depth=135, random_state=42)
tree_clf.fit(X, y)
```

```python
import warnings

warnings.filterwarnings(action='ignore')
num = 0
for j in range(len(catDisease)):
  y_pred = tree_clf.predict(np.array(catDisease.iloc[j][3:]).reshape(1, -1))
  if catDisease.iloc[j][1] != y_pred:
    num+=1
print(num)
```

```python
x_pred = np.array(catDisease.iloc[0][3:]).reshape(1, -1)
x_pred[0] = 0
print(len(x_pred[0]))
for i in range(len(x_pred[0])):
  if catDisease.columns[i].find('침울') != -1:
    x_pred[0][i-3] = 1
  if catDisease.columns[i].find('졸림') != -1:
    x_pred[0][i-3] = 1

print(x_pred)
print()
y_pred = tree_clf.predict(x_pred)
print(y_pred)
```
