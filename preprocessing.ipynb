```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

# 데이터프레임을 읽어옴
petDisease = pd.read_csv('pet.csv',encoding='cp949')
```

```python
# 데이터프레임 출력
petDisease
```

```python
# 데이터프레임의 데이터타입과 값이 없는 행의 개수를 출력
print(petDisease.dtypes)
print(petDisease.isnull().sum())
```

```python
# 필요없는 열 삭제
petDisease = petDisease.drop(['참고문헌','병리소견','정의','발병 및 역학','진단','기타','발병기전'],axis=1)
```

```python
petDisease.head(5)
```

```python
petDisease = petDisease.drop(['치료','예방'],axis=1)
```

```python
petDisease = petDisease.drop(['원인','감별진단','보호자에 대한 조언'],axis=1)
```

```python
petDisease.head(3)
```

```python
# '축종', '주요증상' 열에서 값을 가지지 않는 열을 삭제
petDisease = petDisease.dropna(subset=['축종'])
petDisease = petDisease.dropna(subset=['주요증상'])

print(petDisease.isnull().sum())
```

```python
# 개/고양이로 나누는 프로세스
# 개를 포함하는 행만 추출해서 petDiseaseDog 데이터프레임에 저장
petDiseaseDog = petDisease.loc[petDisease.index[petDisease['축종'].str.contains('개')].tolist(),:]
```

```python
petDiseaseDog
```

```python
petDiseaseDog = petDiseaseDog.drop(['축종'],axis=1)
petDiseaseDog
```

```python
petDiseaseCat = petDisease.loc[petDisease.index[petDisease['축종'].str.contains('고양이')].tolist(),:]
petDiseaseCat
```

```python
petDiseaseCat = petDiseaseCat.drop(['축종'],axis=1)
petDiseaseCat
```

```python
petDiseaseCat.to_csv("petDiseaseCat.csv")
petDiseaseDog.to_csv("tepetDiseaseDogst.csv")
```

```python
# 주요증상을 공백과 '\' 문자로 분리하여 토큰화
import nltk

petDiseaseDog = petDiseaseDog.drop(['예후'],axis=1)
petDiseaseCat['주요증상'] = petDiseaseCat['주요증상'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
# 한글과 공백을 제외하고 모두 제거
petDiseaseCat[:10]
```

```python
from konlpy.tag import Twitter

twt = Twitter()

# 불용어 정의
stopwords=['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
all_words = []
satis = []

for n in range(len(petDiseaseCat)):
  text = petDiseaseCat['주요증상'].iloc[n]
  words = twt.pos(text)

  for i, j in words:
    if j == "Noun":
      if i not in stopwords:
        all_words.append(i)
print(all_words)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

```python
c = Counter(all_words)
wc = WordCloud(font_path='malgun.ttf', width=1400, height=1400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)
```
