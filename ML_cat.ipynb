```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

# 데이터프레임을 읽어옴
catDisease = pd.read_csv('petDiseaseCatv3.csv', encoding='utf-8')
```

```python
catDisease
```

```python
print(catDisease.dtypes)
print(catDisease.isnull().sum())
```

```python
catDisease = catDisease.dropna(subset=['질병명'])
catDisease = catDisease.dropna(subset=['주요증상'])
print(catDisease.isnull().sum())
```

```python
catDisease['주요증상'] = catDisease['주요증상'].str.replace(" ","")
catDisease.head(30)
```

```python
catDisease = catDisease.reset_index(drop=True)
catDisease
```

```python
!pip install konlpy
```

```python
from konlpy.tag import Twitter

twt = Twitter()

all_words = []

for n in range(len(catDisease)):
  text = catDisease['주요증상'].iloc[n]
  text = text.replace('|',',')
  text = text.replace('\r','')
  words = text.split(',')
  for word in words:
    if word not in all_words and len(word)>0:
        all_words.append(word)

print(len(all_words),all_words)
```

```python
all_words
```

```python
catDisease['주요증상']
```

```python
for word in all_words:
  catDisease[word] = 0
  for i in range(len(catDisease)):
    text = str(catDisease.loc[i, '주요증상'])
    if text.find(word)!=-1:
      catDisease.loc[i, word] = 1
catDisease['비정상의폐포음']
```

```python
# 파이썬 ≥3.5 필수
import sys
assert sys.version_info >= (3, 5)

# 사이킷런 ≥0.20 필수
import sklearn
assert sklearn.__version__ >= "0.20"

# 공통 모듈 임포트
import numpy as np
import os

# 노트북 실행 결과를 동일하게 유지하기 위해
np.random.seed(42)

# 깔끔한 그래프 출력을 위해
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# 그림을 저장할 위치
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "decision_trees"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("그림 저장:", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

X = catDisease.iloc[:,2:]
y = catDisease['질병명']

tree_clf = DecisionTreeClassifier(max_depth=20, random_state=42)
tree_clf.fit(X, y)
```

```python
import os
from graphviz import Source
from sklearn.tree import export_graphviz

export_graphviz(
        tree_clf,
        out_file=os.path.join(IMAGES_PATH, "iris_tree.dot"),
        feature_names=catDisease.columns[2:4],
        class_names=catDisease['질병명'],
        rounded=True
    )

Source.from_file(os.path.join(IMAGES_PATH, "iris_tree.dot"))
```
