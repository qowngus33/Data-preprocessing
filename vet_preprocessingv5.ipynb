```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

vet = pd.read_csv('fulldata_02_03_01_P_vet_v6.csv',encoding='utf-8')
vet
```

```python
vet = vet.drop(["Unnamed: 0"], axis=1)
vet
```

```python
vet = vet.rename(columns={'Unnamed: 1':'소재지전화'})
vet = vet.rename(columns={'Unnamed: 5':'사업장명'})
vet
```

```python
vet = vet.drop_duplicates()
```

```python
vet = vet.reset_index(drop=True)
vet
```

```python
vet = vet.dropna()
vet = vet.reset_index(drop=True)
vet
```

```python
vet_original = pd.read_csv('fulldata_02_03_01_P_vet.csv',encoding='EUC-KR')
vet_original
```

```python
vet_original = vet_original.drop_duplicates()
vet_original = vet_original.reset_index(drop=True)
vet_original
```

```python
vet['전체주소'] = ""
vet
```

```python
temp = vet_original.index[vet_original['사업장명'] == vet['사업장명'][0]].tolist()
temp
vet_original['도로명전체주소'][temp[0]]
```

```python
for i in range(len(vet)):
  temp = vet_original.index[vet_original['사업장명'] == vet['사업장명'][i]].tolist()
  if len(temp) == 1:
    vet['전체주소'][i] = vet_original['도로명전체주소'][temp[0]]
  elif len(temp) == 0:
    print('what the fuck!!!!!!!!!!!!!!!!!!!!!!!!')
  else:
    for j in range(len(temp)):
      temp_str = str(vet_original['도로명전체주소'][temp[j]])
      if temp_str.find(vet['Unnamed: 2'][i]) != -1 and temp_str.find(vet['Unnamed: 3'][i]) != -1 and temp_str.find(vet['Unnamed: 4'][i]) != -1:
        vet['전체주소'][i] = vet_original['도로명전체주소'][temp[j]]
```
