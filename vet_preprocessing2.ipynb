```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

vet = pd.read_csv('fulldata_02_03_01_P_vet_v2.csv',encoding='utf-8')
vet
```

```python
vet = vet.drop(['Unnamed: 0', 'Unnamed: 32'], axis=1)
```

```python
vet['소재지전화'] = vet['소재지전화'].fillna(value="없음")

vet['소재지전화'] = vet['소재지전화'].str.strip()
vet['도로명전체주소'] = vet['도로명전체주소'].str.strip()
vet['사업장명'] = vet['사업장명'].str.strip()

vet
```

```python
call10 = vet['소재지전화'].str.extract('(\d\d\d-\d\d\d-\d\d\d\d)')
call10 = call10.dropna(subset=[0])
call10
```
