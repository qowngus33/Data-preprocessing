```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd

# 데이터프레임을 읽어옴
cat = pd.read_csv('catNAN.csv',encoding='utf-8')
dog = pd.read_csv('dogNAN.csv',encoding='utf-8')
puppy = pd.read_csv('puppyNAN.csv',encoding='utf-8')

df = pd.concat([cat,dog,puppy])
```

```python
# 중복제거
df = df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df
```

```python
df = df.drop([df.columns[0]],axis=1)
df = df.reset_index()
df
```

```python
import nltk 
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')
```

```python
def split_sentence(df,sympomList):
  setence_list = []
  for i in range(len(df)):
    text = df.loc[i,'제목']
    text = str(text)
    sent_tokens = sent_tokenize(text)
    for sent in sent_tokens:
      sent = str(sent)
      flag = False
      for symptom in sympomList:
        if len(sent)>0 and sent.find(symptom) != -1:
          flag = True
      if flag:
        print(sent)
        setence_list.append(sent)

    # text = cat['질문'][i]
    # sent_tokens = sent_tokenize(text)
    # for sent in sent_tokens:
    #   flag = False
    #   for symptom in sympomList:
    #     if sent.find(symptom) != -1:
    #       flag = True
    #   if flag:
    #     setence_list.append(sent)

  result = []
  for i in range(len(setence_list)):
    temp_list = setence_list[i].split('\n')
    for temp in temp_list:
      flag = False
      for symptom in sympomList:
        if len(temp)>0 and temp.find(symptom) != -1:
          flag = True
      if flag:
        result.append(temp)
      
  return result

def label_sentence(setence_list):
  label_list = []
  for sent in setence_list:
    word_list = sent.split(' ')
    text = ""
    for word in word_list:
      if word.find('고양이') != -1 or word.find('강아지') != -1 or word.find('개') != -1:
        text += "S-ANIMAL "
      else:
        text += "O "
      text = text[0:len(text)]
    label_list.append(text)
  return label_list

def label_intent(setence_list, intent):
  label_list = []
  for sent in setence_list:
    label_list.append(intent)
  return label_list
```

```python
symptomList = ['피부','혹','종기','종양','종괴','뾰루지','부음','붓기','부어','부었','진드기']
question = split_sentence(df,symptomList)
label = label_sentence(question)
intent = label_intent(question,"skin")

save_file = {'question': question, 'label': label}
save_file = pd.DataFrame(save_file)
save_file.to_csv('skin.csv', index = False, encoding='utf-8')

save_file = {'question': question, 'label': intent}
save_file = pd.DataFrame(save_file)
save_file.to_csv('intent.csv', index = False, encoding='utf-8')
```
